**EXPLORATORY DATA ANALYSIS**

Gender

```python
df.columns
```

```python
ax = sns.countplot( x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

*From above graphs we can see most of the buyers are female as well as the amount spent by females is higher*

Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data = sales_age)
```

*From above graphs we can see that most of the buyers are from age group between 26-35 yrs female*

**State**

```python
sales_st = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x = 'State', y = 'Orders', data = sales_st)
```

```python
sales_st = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x = 'State', y = 'Amount', data = sales_st)
```

*From above graphs we can see that most of the order and total sales/amount is from Uttar Pradesh, Maharashtra and Karnataka respectively*

Marital Status

```python
ax = sns.countplot(x = 'Marital_Status', data = df)

sns.set(rc={'figure.figsize':(7,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ms = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,4)})
sns.barplot(x = 'Marital_Status', y = 'Amount', hue = 'Gender', data = sales_ms)
```

*From above graphs we can see that most of the buyers are married women*

Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x = 'Occupation', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y = 'Amount', data = sales_occ)
```

*From the above graph we can see that most of the buyers are from IT, Healthcare and Aviation Sector*

Product Category

```python
sns.set(rc={'figure.figsize':(30,10)})
ax = sns.countplot(x = 'Product_Category', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pc = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc={'figure.figsize':(33,10)})
sns.barplot(x='Product_Category', y = 'Amount', data = sales_pc)
```

*From above graphs we can see that most of the products are sold from Food, Clothing & Apparel and Electronic category*

Top 10 selling product IDs

```python
sales_pi = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID', y = 'Orders', data = sales_pi)
```

**Conclusion**

Married women of age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation Sector are more likely to buy products from Food, Clothing and Electronics category
